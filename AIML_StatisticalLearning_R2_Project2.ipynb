```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_1samp, ttest_ind, mannwhitneyu, levene, shapiro,wilcoxon, ttest_rel
from statsmodels.stats.power import ttest_power
```

```python
data_df = pd.read_csv("data.csv")
data_df.head()
```

##### Find the mean of old scheme and new scheme column.

```python
data_df["Old Scheme (in thousands)"].mean()
```

```python
data_df["New Scheme (in thousands)"].mean()
```

##### Use the five percent significance test over the data to determine the p value to check new scheme has significantly raised outputs?

```python
Old_scheme = data_df["Old Scheme (in thousands)"]
new_scheme = data_df["New Scheme (in thousands)"]
```

```python
# paired t-test: doing two measurments on the same experimental unit
#h0 = diff in mean is 0 (not significantly raised output)
#h1 = diff in mean is greater than 0 (significantly raised output)
t_statistic, p_value = ttest_1samp(new_scheme - Old_scheme, 0)
#print(t_statistic, p_value)
print ("paired t-test p-value=", p_value)
```

##### What conclusion does the test (p-value) lead to?

since the value of p_value is greater than 0.05, we accept NULL hypothesis; new scheme has not significantly raised outputs.

##### What is the p- value of the hypothesis test if we test for a difference of  $ 5000?

```python
t_statistic, p_value = ttest_1samp(new_scheme - Old_scheme, 5000)
#print(t_statistic, p_value)
print ("paired t-test p-value=", p_value)
```
